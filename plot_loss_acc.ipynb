```python
import matplotlib.pylab as plt
import os
```

```python
workspace = 'loss'
f_txt = 'train_loss_2c.txt'
v_txt = 'val_loss_2c.txt'
```

```python
work_path = os.path.join(workspace,f_txt)
work_path_v = os.path.join(workspace,v_txt)
```

```python
def show_(work_path,work_path_v,epoch=200):
    train_loss_c = []
    train_acc_c = []
    train_loss_s = []
    train_acc_s = []
    
    val_loss_c = []
    val_acc_c = []
    val_loss_s = []
    val_acc_s = []
    
    with open(work_path) as f:
        for line in f.readlines():
#             print(line.split(' '))
#             print(line.strip().split(' ')[3])
            train_loss_c.append(float(line.strip().split(' ')[3].strip()))
            train_acc_c.append(float(line.strip().split(' ')[5].strip()))
            
            train_loss_s.append(float(line.strip().split(' ')[7]))
            train_acc_s.append(float(line.strip().split(' ')[9].strip()))
     
    with open(work_path_v) as f:
        for line in f.readlines():
#             print(line.split(' '))
#             print(line.strip().split(' ')[3])
            val_loss_c.append(float(line.strip().split(' ')[3].strip()))
            val_acc_c.append(float(line.strip().split(' ')[5].strip()))
            
            val_loss_s.append(float(line.strip().split(' ')[7]))
            val_acc_s.append(float(line.strip().split(' ')[9].strip()))
    #
    print('cala  loss:')
    plt.plot(range(epoch),train_loss_c,'r')
    plt.plot(range(epoch),val_loss_c,'b')
    plt.title('c_loss')
    plt.show()
    
    print('soft  loss:')
    plt.plot(range(epoch),train_loss_s,'r')
    plt.plot(range(epoch),val_loss_s,'b')
    plt.title('s_loss')
    plt.show()
    
    print('cala  acc:')
    plt.plot(range(epoch),train_acc_c,'r')
    plt.plot(range(epoch),val_acc_c,'b')
    plt.title('c_acc')
    plt.show()
    
    print('soft acc:')
    plt.plot(range(epoch),train_acc_s,'r')
    plt.plot(range(epoch),val_acc_s,'b')
    plt.title('s_acc')
    plt.show()
```

```python
show_(work_path,work_path_v,112)
```

```python
show_(work_path,work_path_v,401)
```
